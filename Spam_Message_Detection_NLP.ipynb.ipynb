```python
import pandas as pd
import nltk
```

```python
D = pd.read_csv("spam.csv")
D.head(5)
```

```python
D['v2']=D['v2'].apply(lambda x:"".join(x.lower()for x in x.split()))
D['v2']
```

```python
D['v2']=D['v2'].str.replace('[^\w\s]',"")
D['v2']       
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
D['v2']=D['v2'].apply(lambda x:"".join(x for x in x.split()if x not in stop))
D['v2']
```

```python
from nltk.stem import PorterStemmer
st =PorterStemmer()
D['v2'][:5].apply(lambda x:"".join([st.stem(word)for word in x.split()]))
```

```python
from textblob import Word
D['v2']=D['v2'].apply(lambda x:"".join([Word(word).lemmatize()for word in x.split()]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidVectorizer()
vectorizer.fit(D)
```
